```python
import jyrobot
```

```python
world = jyrobot.load_world("duck-scribbler-no-walls")
```

```python
world.watch()
```

```python
world.reset()
world.robots["White"].move(0, 1)
world.robots["Red"].move(1, 0)
```

```python
def control(world):
    for robot in world.robots:
        if robot.stalled:
            robot.reverse()
        if robot.name == "Red":
            pic = robot["camera"].take_picture()
            world.display(pic)
```

```python
%%time
world.run(control)
```
